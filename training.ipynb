```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import os

SAVE_DIR = "./Processed_Data/"
base_name = "RealDrum01_10#MIX" 
# 1. Load the processed .npy file
x_data = np.load(os.path.join(SAVE_DIR,f"{base_name}_X.npy"))
y_data = np.load(os.path.join(SAVE_DIR,f"{base_name}_y.npy"))

print(f"Spectrogram Shape: {x_data.shape}") # Expecting (Total_Frames, 128)
print(f"Label Shape: {y_data.shape}")       # Expecting (Total_Frames, 3)

# 2. Setup the plot
plt.figure(figsize=(12, 8))

# --- Top Plot: The Spectrogram ---
plt.subplot(2, 1, 1)
# We transpose it back (.T) so Time is on the X-axis and Bins are on the Y-axis
librosa.display.specshow(x_data.T, sr=44100, hop_length=512, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title("Mel-Spectrogram (Input X)")

# --- Bottom Plot: The Ground Truth Labels ---
plt.subplot(2, 1, 2)
plt.plot(y_data[:, 0], label='Kick (KD)', color='red', alpha=0.7)
plt.plot(y_data[:, 1], label='Snare (SD)', color='blue', alpha=0.7)
plt.plot(y_data[:, 2], label='Hi-Hat (HH)', color='green', alpha=0.7)
plt.xlabel("Frame Index")
plt.ylabel("Hit (0 or 1)")
plt.legend(loc='upper right')
plt.title("Ground Truth Annotations (Target Y)")

plt.tight_layout()
plt.show()
```
